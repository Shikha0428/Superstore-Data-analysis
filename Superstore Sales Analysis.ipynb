```python
# importing library
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
# loading superstore sales dataset
df = pd.read_csv("superstore_final_dataset.csv")
```

```python
# Looking at top 5 rows of superstore dataset
df.head()
```

```python
# Looking at information of superstore dataset
df.info()
```

```python
# Looking at the dimension
df.shape
```

```python
# Find the duplicate rows except Row_ID
duplicateRows = df[df.duplicated(subset=df.columns.difference(['Row_ID']))]
```

```python
duplicateRows
```

```python
# To drop the 1 duplicated row
df = df.drop_duplicates(subset=df.columns.difference(['Row_ID']))
```

```python
df
```

```python
# To find out the rows with missing null values
null_data = df[df.isnull().any(axis=1)]
```

```python
null_data
```

```python
# To remove rows with null values.
df = df.dropna()
```

```python
df
```

```python
# To change the Date datatype from Object to datetime
df['Order_Date'] = pd.to_datetime(df['Order_Date'], yearfirst=True)
df['Ship_Date'] = pd.to_datetime(df['Ship_Date'], yearfirst=True)
```

```python
df
```

```python
df.dtypes
```

```python
# To Check whether any record having Order_Date is greater than Ship_Date
df = df[df['Ship_Date'] >= df['Order_Date']]
```
